```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import(PowerNorm,LogNorm)
from astropy.io import fits
import os
import glob
```

# Bias

```python
bias_filenames = np.array(glob.glob('./Final_Data/Bias/*.fits'))
bias_frames = []
for filename in bias_filenames:
    bias_frames.append(fits.getdata(filename))
bias_filenames
```

```python
plt.figure(figsize=(20,20))

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(bias_frames[i], cmap='gray', origin='lower', norm=LogNorm())

plt.show()
```

```python
bias = np.median(bias_frames, axis=0)

fig = plt.figure(figsize=(10,10))
plt.imshow(bias, origin='lower', cmap='gray', norm=LogNorm())
plt.axis('off')
fig.savefig('./Output/Figures/Bias.pdf', bbox_inches='tight')
plt.show()
```

```python
hdu_bias = fits.PrimaryHDU(bias)
hdul_bias = fits.HDUList([hdu_bias])
hdu_bias.writeto('./Output/Bias.fits', overwrite='True')
```
